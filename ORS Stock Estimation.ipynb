```r
# Install necessary packages
# install.packages("dplyr")

library(dplyr)

# Read the file 
df <- read.csv("data.csv", check.names = FALSE)
```

```r
library(dplyr)


# 1) Incidence columns
inc_cols <- grep("^Diarrhoea <5 y incidence rate", names(df), value = TRUE)

df <- df %>%
  mutate(avg_incidence = rowMeans(across(all_of(inc_cols)), na.rm = TRUE))

df_inc <- df %>%
  select(organisationunitname, avg_incidence) %>%
  arrange(desc(avg_incidence))

head(df_inc)
top_district <- df_inc$organisationunitname[1]

# 2) Numerator columns (episodes)
dys_cols <- grep("^Diarrhoea with blood \\(Dysentery\\) new", names(df), value = TRUE)
sev_cols <- grep("^Diarrhoea with severe dehydration follow-up", names(df), value = TRUE)

top_row <- df %>% filter(organisationunitname == top_district)

total_episodes_12m <- sum(top_row[, dys_cols], top_row[, sev_cols], na.rm = TRUE)
avg_episodes_month <- total_episodes_12m / 12

# 3) ORS calculation
ors_per_episode <- 3 * 5  # 3 packets/day * 5 days
min_ors_month <- avg_episodes_month * ors_per_episode

buffer_factor <- 1.2
recommended_ors <- min_ors_month * buffer_factor

cat("District with highest average incidence:", top_district, "\n")
cat("Average diarrhoea episodes per month:", round(avg_episodes_month, 2), "\n")
cat("Minimum ORS packets per month (no buffer):", round(min_ors_month), "\n")
cat("Recommended ORS packets per month (20% buffer):", ceiling(recommended_ors), "\n")
```
